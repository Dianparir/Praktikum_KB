#### Substitusi

```python
imputer_mean = SimpleImputer(strategy='mean')
imputer_median = SimpleImputer(strategy='median')
imputer_most_frequent = SimpleImputer(strategy='most_frequent')
```

```python
ds_rusak_cleaned['webtoon_id'] = imputer_mean.fit_transform(ds_rusak[['webtoon_id']])
ds_rusak_cleaned['episodes'] = imputer_median.fit_transform(ds_rusak[['episodes']])
ds_rusak_cleaned['episodes'] = imputer_most_frequent.fit_transform(ds_rusak[['episodes']])
```

```python
ds_rusak.isna().sum()
```

```python
ds_pembanding = pd.DataFrame(
{
    "webtoon_id_original":ds_rusak['webtoon_id'],
    "webtoon_id_cleaned":ds_rusak_cleaned['webtoon_id'],
     "episodes_original":ds_rusak['episodes'],
    "episodes_cleaned":ds_rusak_cleaned['episodes'],
     "grade_original":ds_rusak['grade'],
    "grade_cleaned":ds_rusak_cleaned['grade']
}
)
```

### One Hot Encoding

```python
ds_one_hot_encoding = ds.copy()
```

```python
ds.head(10)
```

```python
ds['genre']
```

```python
# buat data frame baru
ds_OHE = ds_one_hot_encoding.copy()
```

```python
# buat tools
encoder_OH = OneHotEncoder(sparse=False)
```

```python
# encoding label atribut dari kategorikal ke biner
temp_dataset = encoder_OH.fit_transform(ds_one_hot_encoding[['genre']]).astype('int32')
```

```python
# format ulang dataset menjadi dataframe
temp_dataset = pd.DataFrame(temp_dataset)
```

```python
# drop atribut label sebelumnya
ds_OHE.drop('genre', axis=1, inplace=True)
```

```python
# tambahkan atribut label baru
ds_OHE = ds_OHE.join(temp_dataset)
```

```python
# ganti nama label atribut 'genre'
ds_OHE.rename(columns={0:'Romance', 1:'Fantasy', 2:'Horror', 3:'Drama', 4:'Action', 5:'Thriller', 6:'Sci-fi', 7:'Historical',
                       8:'Supernatural', 9:'Comedy', 10:'Slice of life', 11:'Mystery', 12:'Heartwarming', 13:'Superhero', 14:'Sports',
                       15:'Informative'}, inplace=True)
```
