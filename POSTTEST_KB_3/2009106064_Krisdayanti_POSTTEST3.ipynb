## Grafik Batang

```python
y = dataset['Location ISO Code']
x = dataset['New Cases']
```

```python
plt.bar(x, y, align='center', width = 200, bottom = 0.5)
plt.title('COVID-19 Indonesia Dataset')
plt.xlabel('Location ISO Code')
plt.ylabel('New Cases')
plt.show()
```

Grafik di atas menunjukkan bahwa jumlah kasus baru COVID-19 lebih tinggi di daerah Gorontalo (ID-GO)

## Grafik Pie

```python
dataset = dataset.dropna(subset=['Island'])
```

```python
data_pie = dataset['Island'].value_counts().rename_axis('Island').reset_index(name='Id_count')
```

```python
plt.figure(figsize=(10,10))
plt.pie(data_pie.Id_count, labels=data_pie.Island, startangle=90,
autopct='%.1f%%')
plt.title('Persentase Penyebaran COVID-19 di setiap Pulau')
plt.show()
```

Dari grafik di atas bisa disimpulkan bahwa pulau Sumatera memiliki tingkat Persentase Penyebaran COVID-19 lebih tinggi dibadingkan dengan pulau Papua dan Maluku yang memiliki persentase lebih rendah.

```python
sns.scatterplot(data=dataset,
                x='Total Cases',
                y='Location ISO Code')
plt.title('Total Kasus COVID-19 di setiap Lokasi')
plt.show()
```

```python
sns.scatterplot(data=dataset,
                x='Total Deaths',
                y='Island')
plt.title('Total Kematian penduduk akibat COVID-19 di setiap Lokasi')
plt.show()
```

```python
sns.boxplot(dataset['Total Recovered'])
plt.title('Grafik Boxplot Total Penduduk yang pulih dari COVID-19 di setiap Lokasi')
plt.show()
```
